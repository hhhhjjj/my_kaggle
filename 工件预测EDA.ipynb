```python
import pandas as pd
pd.set_option('display.max_column',30)
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from pylab import rcParams ##matplotlib
rcParams['figure.figsize'] = 12, 8
```

```python
train = pd.read_csv('first_round_training_data.csv')[['Parameter'+str(i) for i in range(1,11)]+['Quality_label']]
# 读取初赛需要的特征
test = pd.read_csv('first_round_testing_data.csv')
```

```python
train['Quality_label'].value_counts().plot(kind='bar')
# pycharm显示jupyter notebook的图需要在pycharm上面看
plt.show()
# 要画出来需要加show
```

```python
train['label'] = train['Quality_label'].map({
    'Excellent':1,
    'Good':2,
    'Pass':3,
    'Fail':4
})
```

```python
train.describe()
# 第一四分位数 (Q1)，又称“较小四分位数”，等于该样本中所有数值由小到大排列后第25%的数字。

# 第二四分位数 (Q2)，又称“中位数”，等于该样本中所有数值由小到大排列后第50%的数字。

# 第三四分位数 (Q3)，又称“较大四分位数”，等于该样本中所有数值由小到大排列后第75%的数字。
```

```python
plt.figure(figsize=(12,8))
correlation = train.corr()
mask = np.zeros_like(correlation)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(data=correlation.round(3),mask=mask,vmin=-0.1,vmax=0.1,annot=True, linewidths=.1)
plt.show()
# 看变量之间的相关性
```

```python
train.nunique().plot(kind='bar')
# 判断是否是类别变量
plt.show()
# 如果类别多的就不是，类别少的就可能是类别变量
```

```python
# 查看各个参数label的分布
for i in ['Parameter'+str(i) for i in range(1,11)]:
    sns.scatterplot(x='Quality_label', y=i, data=train)
    plt.title(i)
    plt.show()
```

```python
# 后面几个参数明显是类别变量，我们就看后面几个参数的分布
# 训练集和测试集对比
for i in ['Parameter'+str(i) for i in range(5,11)]:
    train[i].value_counts().plot(kind='bar',alpha=0.5, label='Train')
    test[i].value_counts().plot(kind='bar',alpha=0.5,color='r', label='Test')
    plt.title(i)
    plt.legend()
    plt.show()
```

```python
# 核密度估计(kernel density estimation)是在概率论中用来估计未知的密度函数，属于非参数检验方法之一。
# 通过核密度估计图可以比较直观的看出数据样本本身的分布特征。
for i in ['Parameter'+str(i) for i in range(1,11)]:
    sns.kdeplot(train[i], label='Train',alpha=0.8)
    sns.kdeplot(test[i], label='Test',alpha=0.8)
    plt.title(i)
    plt.show()
```

```python
for i in ['Parameter'+str(i) for i in range(1,11)]:
    plt.figure(figsize=(14,8))
#     这个就是画散点图看数据分布
    plt.scatter(x=range(len(train)), y=train[i], label='Train')
    plt.scatter(x=range(len(test),len(test)*2), y=test[i], label='Test')
    plt.title(i)
    plt.legend()
    plt.show()
```

```python
# 这个对数图有什么用？
for i in ['Parameter'+str(i) for i in range(1,11)]:
#     sorted() 函数对所有可迭代的对象进行排序操作
    plt.scatter(x=range(len(train)), y=sorted(np.log(train[i])), label='Train')
    plt.scatter(x=range(len(test)), y=sorted(np.log(test[i])), label='Test')
    plt.title(i+'--log')
    plt.legend()
    plt.show()
    
    sm.qqplot(np.log(train[i]), line='s', label='Train')
    plt.legend()
    plt.title(i+'--QQ Plot')
    plt.show()
```

```python
import pandas as pd
pd.set_option('display.max_column',30)
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from pylab import rcParams ##matplotlib
rcParams['figure.figsize'] = 12, 8

import warnings 
warnings.filterwarnings('ignore')

from pyecharts import options as opts
from pyecharts.charts import Bar, Line, Pie, Map, Page, ThemeRiver, Timeline, Grid
from pyecharts.commons.utils import JsCode
```

```python
train['Parameter1'] = np.log(train['Parameter1']+1e-05)
train['Parameter2'] = np.log(train['Parameter2']+1e-05)
train['Parameter3'] = np.log(train['Parameter3']+1e-05)
train['Parameter4'] = np.log(train['Parameter4']+1e-05)
```

```python
def parameter_label(data, title, max_, bin_) -> Bar:
    c = (
        Bar()
        .add_xaxis(list(range(bin_)))
        .add_yaxis("Excellent", list(data[data['Quality_label']=='Excellent']['count']), stack="stack1")
        .add_yaxis("Good", list(data[data['Quality_label']=='Good']['count']), stack="stack1")
        .add_yaxis("Pass", list(data[data['Quality_label']=='Pass']['count']), stack="stack1")
        .add_yaxis("Fail", list(data[data['Quality_label']=='Fail']['count']), stack="stack1")
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(
            title_opts=opts.TitleOpts(title=title,subtitle='James He'),
            toolbox_opts=opts.ToolboxOpts(),
            yaxis_opts=opts.AxisOpts(
                max_= max_,
            ),
        )
    )
    return c
```

```python
bin_ = 30

train['Parameter1_bin'] = pd.cut(train['Parameter1'], bins=bin_, labels=list(range(bin_)))

data = train.groupby(['Parameter1_bin','Quality_label'],as_index=False)['Parameter1'].count().rename(
    columns={'Parameter1':'count'})

c = parameter_label(data, 'Parameter1', 600, bin_)
c.render_notebook()
# 前面四个特征都是正态分布，对预测没什么影响
```
